<h2>F(R) Gravity Theoretical Framework</h2>

The $F(R)$ gravity action we shall consider in the present notebook, has the following
form,
$$ \mathcal{S}=\int d^4x\sqrt{-g}\left[
\frac{1}{2\kappa^2}F(R)+\mathcal{L}_m
\right]\, ,
$$
where $\kappa^2=\frac{1}{8\pi G}=\frac{1}{M_p^2}$, with $G$ being
Newton's gravitational constant and $M_p$ being the reduced Planck
mass. Also the Lagrangian $\mathcal{L}_m$ contains all the perfect
fluids that we be assumed to be present in the theory under study, which are relativistic matter (radiation) and and non-relativistic non-baryonic and non interacting matter (cold-dark-matter). 

The form of
the $F(R)$ gravity which we shall assume for has the following form:
<b>$$
F(R)=R+f(R)\, ,
$$</b>
and for the programming we shall assume that the $F(R)$ gravity has exactly the above form. As an example we shall study the following modification of Einstein's GR,
$$
f(R)=\frac{1}{M^2}R^2+\lambda R e^{\frac{-5.1 \Lambda}{R}}+5.1\lambda \Lambda- \frac{\Lambda}{0.2}
\Big{(}\frac{R}{3m_s^2}\Big{)}^{0.1}\, ,
$$
with $m_s$ being
$m_s^2=\frac{\kappa^2\rho_m^{(0)}}{3}$, where $\rho_m^{(0)}$ is the current dark energy density. The late-time phenomenology of this $F(R)$ gravity is studied in detail in 
<b>New paper to appear, V.K. Oikonomou</b>
The parameter
$\Lambda $ has mass dimensions $[m]^2$, and will be assumed to take values of the order of the cosmological constant. The
parameter $M$ must be approximately $M= 1.5\times
10^{-5}\left(\frac{N}{50}\right)^{-1}M_p$ for early time
phenomenological reasoning, with $N$ denoting the
$e$-foldings number. For the whole study we shall assume that the background metric is a flat Friedmann-Robertson-Walker
(FRW) metric, with line element,
$$ds^2 = - dt^2 + a(t)^2 \sum_{i=1,2,3}
\left(dx^i\right)^2\, ,
$$
So by using the metric formalism, by varying the gravitational $F(R)$ action with respect to the metric tensor
we obtain the gravitational equations of motion,
$$
3 H^2F_R=\frac{RF_R-F}{2}-3H\dot{F}_R+\kappa^2\left(
\rho_r+\rho_m\right)\, ,
$$
$$
-2\dot{H}F=\ddot{F}_R-H\dot{F}_R
+\frac{4\kappa^2}{3}\rho_r\, ,
$$
where $F_R=\frac{\partial F}{\partial R}$. In addition, the
``dot'' denotes differentiation with respect to the cosmic time
$t$, and in addition, we assumed that radiation and CDM are present, so $p_r=\frac{1}{3}\rho_r$, and $\rho_m$ stands for the energy density of the cold dark matter perfect fluid.



<h2>Late-time F(R) Gravity Theoretical Framework, Using the Redshift as a Dynamical Variable</h2>


The gravitational equations of motion can be cast in a
Einstein gravity similar form for a flat FRW spacetime,
$$
3H^2=\kappa^2\rho_{tot}\, ,
$$
$$
-2\dot{H}=\kappa^2(\rho_{tot}+P_{tot})\, ,
$$
where $\rho_{tot}=\rho_{m}+\rho_{DE}+\rho_r$ is the total energy
density of the cosmological fluid and $P_{tot}=P_r+P_{DE}$ is
the total pressure. In our case, the total fluid consists
of radiation with energy density $\rho_r$, of
dark matter with energy density $\rho_{m}$, and of the geometric fluid
$\rho_{DE}$ which at late-times will effective drive the dark energy era, and its energy density is,
$$
\kappa^2\rho_{DE}=\frac{F_R R-F}{2}+3H^2(1-F_R)-3H\dot{F}_R\, .
$$
The corresponding pressures for radiation and the gemoetric fluid are,
$$P_r=\frac{1}{3}\rho_r$$
$$
\kappa^2 P_{DE}=\ddot{F}_R-H\dot{F}_R+2\dot{H}(F_R-1)-\kappa^2 \rho_{DE}\, .
$$
The fluids are perfect, so they satisfy the continuity equation,
$$
 \dot{\rho}_m+3H(\rho_m+P_m)=0\, , 
$$
$$
\dot{\rho}_r+3H(\rho_r+P_r)=0\, , 
$$
$$
\dot{\rho}_{DE}+3H(\rho_{DE}+P_{DE})=0\, .
$$
Aiming for the late-time study, we shall convert the Friedmann equation
in terms of a new convenient variable, the redshift $z$ defined as follows,
$$
1+z=\frac{1}{a}\, ,
$$
and we assumed that the scale factor at present day is 
(at $z=0$) is equal to one. Also, we introduce the
statefinder function $y_H(z)$ which effectivelly will quantify our study
<i>(0705.1158,1207.1009,2001.06830,new)</i>,
defined in the following way,
$$
y_H(z)=\frac{\rho_{DE}}{\rho^{(0)}_m}\, ,
$$
where $\rho^{(0)}_m$ is the energy density of cold
dark matter at present. By using the first Friedman equation,
the statefinder function $y_H(z)$ can be written as follows,
$$
y_H(z)=\frac{3H^2}{\kappa^2\rho^{(0)}_m}-\frac{\rho_M}{\rho^{(0)}_m}-\frac{\rho_r}{\rho^{(0)}_m}\,
.
$$
The radiation energy density as a function of the redshift, scales as
$\rho_r=\rho_r^{(0)}a^{-4}$, where $\rho_r^{(0)}$ is the  value of the radiation energy density at present
time, so
$\frac{\rho_r}{\rho^{(0)}_m}=\chi (1+z)^4$, where
$\chi=\frac{\rho^{(0)}_r}{\rho^{(0)}_m}\simeq 3.1\times 10^{-4}$.


Now upon
substituting $\frac{\rho_r}{\rho^{(0)}_m}=\chi (1+z)^4$, the statefinder
function $y_H(z)$ becomes,
$$
y_H(z)=\frac{H^2}{m_s^2}-(1+z)^{3}-\chi (1+z)^4\, .
$$
where the parameter
$m_s^2=\frac{\kappa^2\rho^{(0)}_m}{3}=H_0\Omega_c=1.37201\times
10^{-67}$eV$^2$. Now we can express the gravitational equations of motion as a function of the statefinder function
$y_H(z)$, so it can be shown (1207.1009,2001.06830,new) that the resulting equation is,
$$
\frac{d^2y_H(z)}{d z^2}+\mathcal{J}_1\frac{d y_H(z)}{d z}+\mathcal{J}_2y_H(z)+\mathcal{J}_3=0\,
,
$$
where the functions $\mathcal{J}_1$, $\mathcal{J}_2$ and $\mathcal{J}_3$ are defined in the following way,
$$
\mathcal{J}_1=\frac{1}{z+1}\left(
-3-\frac{1-F_R}{\left(y_H(z)+(z+1)^3+\chi (1+z)^4\right) 6
m_s^2F_{RR}} \right)\, ,
$$
$$
\mathcal{J}_2=\frac{1}{(z+1)^2}\left(
\frac{2-F_R}{\left(y_H(z)+(z+1)^3+\chi (1+z)^4\right) 3
m_s^2F_{RR}} \right)\, ,
$$
$$
\mathcal{J}_3=-3(z+1)-\frac{\left(1-F_R
\right)\Big{(}(z+1)^3+2\chi (1+z)^4
\Big{)}+\frac{R-F}{3m_s^2}}{(1+z)^2\Big{(}y_H(z)+(1+z)^3+\chi(1+z)^4\Big{)}6m_s^2F_{RR}}\,
$$
where $F_{RR}=\frac{\partial^2 F}{\partial R^2}$. This differential equation is the starting point of our numerical analysis, so by using appropriate initial conditions, which are physically motivated, we aim to solve the above numerically. The initial conditions will be appropriate for describing the late stages of the matter domination era, and we shall focus on the redshif interval $z=[z_i,z_f]$ with $z_i=0$ and $z_f=10$.

The Ricci scalar as a  function of $y_H(z)$ can be written in the following way,
$$
R(z)=3m_s^2\left( 4y_H(z)-(z+1)\frac{d y_H(z)}{d
z}+(z+1)^3\right)\, .
$$
The initial conditions for the problem at hand, will be characteristic of the last stages of the matter domination epoch, so at 
$z_f=10$ these are (1207.1009,2001.06830,new),
$$
y_H(z_f)=\frac{\Lambda}{3m_s^2}\left(
1+\frac{(1+z_f)}{1000}\right)\, , \,\,\,\frac{d y_H(z)}{d
z}\Big{|}_{z=z_f}=\frac{1}{1000}\frac{\Lambda}{3m_s^2}\, ,
$$

For our study, we shall be interested in several physical quantities of cosmological interest, and we shall compare these to the corresponding  $\Lambda$CDM model ones at $z=0$, or we shall confront these with the latest Planck constraints on cosmological parameters (1807.06209).

Particularly we shall be interested in:

The dark energy EoS
parameter $\omega_{DE}=\frac{P_{DE}}{\rho_{DE}}$, can be expressed in
terms of the statefinder function $y_H$ in the following way,
$$
\omega_{DE}(z)=-1+\frac{1}{3}(z+1)\frac{1}{y_H(z)}\frac{d y_H(z)}{d
z}\, ,
$$


The dark energy density parameter $\Omega_{DE}=\frac{\rho_{DE}}{\rho_{tot}}$, which in terms of $y_H(z)$ is ,
$$
\Omega_{DE}(z)=\frac{y_H(z)}{y_H(z)+(z+1)^3+\chi (z+1)^4}\, .
$$

Also the
deceleration parameter $q$, is defined as follows,
$$
q=-1+\frac{1+z}{H(z)}\frac{d H(z)}{dz}\, ,
$$
There exist several other statefinder quantities, but in this python notebook we shall be interested only in the deceleration parameter.

<h2>Programming Tips prior Presenting the Code: The solve_ivp module of Python 3 and its Characteristics</h2>

The solve_ivp (stands for initial value porblem), works only for differential equations of the form:
$$\frac{d y}{dt}=f(t,y)$$
where $y$ is a vector of the form

$y=[y_1,y_2,y_3,...,y_n]$ and f is a n-dimensional vector $f=[f_1,f_2,f_3,...,f_n]$. This can easily be done in many cases of even having systems of second order differential equations. The initial conditions are

$$y(t_0)=y_0$$ with $y_0$ an n-dimensional constant vector.

Now an important comment: The n-dimensional function $f(t,x)$ must be defined in exactly the form $f(t,y)$, with the dynamic variable t being on the left, and the variable $y$ depends on the variable t.

The solv_ivp function of scipy module can solve numerically a system of first order differential equations easily by using highly advanced numerical techniques, like LSODA, BDF, RK45, Radau and similar.

Also it is important to note that it's syntax can only integrate forward in the dynamical variable, in our case the redshift. So we need to amend the differential equation which we shall solve, and also amend all the related quantities to tailor these in order for the solve_ivp function to be able to perform the integration.

We remind here the differential equation we need to solve is:
$$
F_{RR}\frac{d^2y_H(z)}{d z^2}+J_1\frac{d y_H(z)}{d z}+J_2y_H(z)+J_3=0\,
,
$$
where the functions $J_1$, $J_2$ and $J_3$ are defined as follows,
$$
J_1=a(z)\left(
-3F_{RR}-\frac{1-F_R}{\left(y_H(z)+a(z)^{-3}+\chi a^{-4}\right) 6
m_s^2} \right)\, ,
$$
$$
J_2=a(z)^{-2}\left(
\frac{(2-F_R)}{\left(y_H(z)+a(z)^{-3}+\chi a(z)^{-4}\right) 3
m_s^2} \right)\, ,
$$
$$
J_3=-3a(z)^{-1}F_{RR}-\frac{\left(1-F_R
\right)\Big{(}a(z)^{-3}+2\chi a(z)^{-4}
\Big{)}+\frac{R-F}{3m_s^2}}{a(z)^{-2}\Big{(}y_H(z)+a(z)^{-3}+\chi a(z)^{-4}\Big{)}6m_s^2}\,
$$
which is obviously second order, and the initial conditions are:
$$
y_H(z_f)=\frac{\Lambda}{3m_s^2}\left(
1+\frac{1}{1000}(1+z_f)\right)\, , \,\,\,\frac{d y_H(z)}{d
z}\Big{|}_{z=z_f}=\frac{1}{1000}\frac{\Lambda}{3m_s^2}\, ,
$$
where $z_f$ is the final redshift. The aim now is to transform the above second order initial value problem to a initial value problem corresponding to a system of first order differential equations, and also amend appropriately the resulting equations in order to address the backwards integration issue.


Firstly, let us make the second order equation, first order, and this can be done by introducing another variable $Y=\frac{dy_H(z)}{d z}$, so the second order differential equation is transformed to the following set of first order ODEs:
$$
\frac{dy_H(z)}{d z}=Y
$$

$$
\frac{dY}{d z}=-\frac{1}{F_{RR}}(J_1\frac{d y_H(z)}{d z}+J_2y_H(z)+J_3)\,
,
$$
and now the set of differential equations has the form,
$$\frac{d y}{dt}=f(t,y)$$
where $t$ is the redshift variable $z$, is the vector $y=(y_H,Y)$ and $f$ is the vector:

$$f=(Y,-\frac{1}{F_{RR}}(J_1\frac{d y_H(z)}{d z}+J_2y_H(z)+J_3))$$

and the initial conditions at the final redshift $z_f=10$ (which in practise is the starting point of our integration):

$$$$

<b>Important notation notice:
When programming, t will be identical to the redshift, but this is just a notation issue for the solve_ivp, the whole system is structured for the redshift to be the dynamical variable.</b>

Now for the backward integration, we shall make the following transformation in the differential equations, and all the related physical quantities, namely,
$$z'=10-z$$
so when $z=0$ in the real world, in the transformed system this corresponds to $z'=10$, and when $z=10$ in the real world, this corresponds to $z'=0$. We wish to integrate from $z=10$ to $z=0$ in the real world, hence this is a backward integration. But in the transformed system, the integration from $z=10$ to $z=0$ in the real world corresponds to $z'=0$ to $z'=10$, so the backward integration is transformed to an ordinary forward integration in the transformed system.

Now let us proceed with the transformation of all the equations to the new system with redshift $z'$ and of the initial conditions. We have the following rules:

$$\frac{d}{dz}=-\frac{d}{dz'}$$


so the second order differential equation becomes,

$$
F_{RR}\frac{d^2y_H(z')}{d z'^2}-J_1\frac{d y_H(z')}{d z'}+J_2y_H(z')+J_3=0\,
,
$$

Now setting: 


$$
\frac{dy_H(z')}{d z'}=Y
$$
the second order differential equation is transformed to a set of two first order differential equations:
$$
\frac{dy_H(z')}{d z'}=Y
$$
$$
\frac{dY}{d z'}=-\frac{1}{F_{RR}}(-J_1\frac{d y_H(z')}{d z}+J_2y_H(z')+J_3)\,
,
$$
while the parameters $J_1$, $J_2$ and $J_3$ become,
$$
J_1=a(z')\left(
-3F_{RR}-\frac{1-F_R}{\left(y_H(z')+a(z')^{-3}+\chi a(z')^{-4}\right) 6
m_s^2} \right)\, ,
$$
$$
J_2=a(z')^{-2}\left(
\frac{(2-F_R)}{\left(y_H(z')+a(z')^{-3}+\chi a(z')^{-4}\right) 3
m_s^2} \right)\, ,
$$
$$
J_3=-3a(z')^{-1}F_{RR}-\frac{\left(1-F_R
\right)\Big{(}a(z')^{-3}+2\chi a(z')^{-4}
\Big{)}+\frac{R-F}{3m_s^2}}{a(z')^{-2}\Big{(}y_H(z')+a(z')^{-3}+\chi a(z')^{-4}\Big{)}6m_s^2}\,
$$
where the redefined scale factor is: 
$$
a(z')=\frac{1}{11-z'}\, .
$$
Now, the initial conditions become in terms of the redefined redshift variable $z'$:
$$
y_H(z'_f)=\frac{\Lambda}{3m_s^2}\left(
1+\frac{1}{1000}(11-z'_f)\right)\, , \,\,\,\frac{d y_H(z')}{d
z'}\Big{|}_{z=z'_f}=-\frac{1}{1000}\frac{\Lambda}{3m_s^2}\, ,
$$
where $z'_f=0$ in our transformed redshift case. 


Also the curvature in the redefined redshift variable $z'$ reads,

$$
R(z')=3m_s^2\left( 4y_H(z')+a(z')^{-1}\frac{d y_H(z')}{d
z'}+a(z')^{-3}\right)\, .
$$

Let us now consider the physical quantities which we shall study in the following:

$$
\Omega_{DE}(z')=\frac{y_H(z')}{y_H(z')+a(z')^{-3}+\chi a(z')^{-4}}\, .
$$
$$
\omega_{DE}(z')=-1-\frac{1}{3}a(z')^{-1}\frac{1}{y_H(z')}\frac{d y_H(z')}{d

$$
q=-1-\frac{11-z'}{H(z')}\frac{d H(z')}{dz'}\, ,
$$

<h2>The Code for F(R) Gravity Dark Energy Phenomenology</h2>

Now the F(R) gravity dark energy code in the presence of non-relativistc non-baryoning non interacting matter and fluids and radiation fluids.

We use the expressions for fR and fRR which are the derivatives of the F(R)=R+f(R) function, the f(R) part. First we shall introduce several constants and variable values tha appear in the model of $f(R)$ under study.

<h3>The Python3 Code and the Numerical Analysis Results</h3>

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.integrate import odeint
from scipy.misc import derivative

#Definition of Hubble rate at present time based on CMB-Planck 2018 data. It is in eV

H0p = (67.4)*(5.067*10**5*10**13*10**9)/(1.592*1.5637*10**24*10**38)

#Definition of Hubble rate at present time based on Cepheids data. It is in eV

H0c = (74)*(5.067*10**5*10**13*10**9)/(1.592*1.5637*10**24*10**38)

#Mass scale (eV)
    
ms = 1.87101*10**(-67)


#Cosmological constant (eV)



Lambda_s= 7.93*1.5*10**(-67)

# the parameter $\chi$, radiation over dark matter present day energy densities.


chi=3.1*10**(-4)



#The initial redshift value of the redefined redshift z'=0, which corresponds to the final value of the original redshift z=10

zfin = 0

#initial conditions
#Notice that we used the redefined redshift z'=10-z, so the value zfin=0 of the redefined redshift corresponds
# to zfin=10 of the initial redshift parameter.

r = Lambda_s/(3*ms)*(1 + (11-zfin)/1000)

phi= -Lambda_s/(3*ms)*1/1000


#Model parameters


M = 3.04375*10**22 # in (eV)

#lambda_s=0.01 #dimensionless
lambda_s=0.9999 #dimensionless

#t=z from now on in order for the solve_ivp function to work.
#Now you can use fR in your programm


def f(t,y):
    yH=y[0]
    Y=y[1]
    #first definition of the curvature for the numerical integration. This must be re-defined after the numerical integration
    #in order to be expressed in terms of the new solution
    a=1/(11-t)
    rs=(chi)*a**(-4)
    R=3*ms*((1/a)*Y + 4*yH + a**(-3))
    #The fR and fRR are taken from above
    f=R**2/M**2+lambda_s*R*np.exp(-Lambda_s*5.1/R)+Lambda_s*5.1*lambda_s-(Lambda_s/0.2)*(R/(ms))**0.1
    fR=5.1*Lambda_s*lambda_s*np.exp(-5.1*Lambda_s/R)/R - 0.5*Lambda_s*(R/ms)**0.1/R + lambda_s*np.exp(-5.1*Lambda_s/R) + 2*R/M**2
    fRR=26.01*Lambda_s**2*lambda_s*np.exp(-5.1*Lambda_s/R)/R**3 + 0.45*Lambda_s*(R/ms)**0.1/R**2 + 2/M**2
    #The fR and fRR are taken from above
    J1=a*(-3*fRR - (1/(yH + a**(-3) + rs))*(-fR/(6*ms)))
    J2=(a**2)*(1/(yH + a**(-3) + rs))*(1 - fR)/(3*ms)
    J3=-3*a**(-1)*fRR - ((a**2)*(-fR*(a**(-3) + 2*rs )- f/(3*ms))/(yH + a**(-3) + rs))*1/(6*ms)
    #we denote the derivative of y_H with respect to the redefined redshift z' as dyHdzr. This is one of the two things 
    #that the f function will give us
    dyHdz=Y
    dYdz=-(1/fRR)*(-J1*Y+J2*yH+J3)
    return np.array([dyHdz,dYdz])
    
# At this point we set the redefined redshift initial and final values: from z=0 to z=10. 
t_span=np.array([0,10])

#And at this point we choose all the intermediate redshift points from z=0 to z=10, using 1000 points. In principle one 
#can choose more intermediate points or even less, it is up to the reader.

time_interval=np.linspace(t_span[0],t_span[1],1000)


# Now we set the initial condition. Since y the outcome of the solve_ivp that will follow, is an array, with y[0] being the 
#function y_H and y[1] being the derivative yH'[z], the initial condition is an array fixing the initial value of yH to be r
#defined previously, and that of yH'[z] to be phi, also defined previously, below the definition of r.


y0=np.array([r,phi])



# Now we can the solve_ivp function of the scipy module:


soly=solve_ivp(f,t_span,y0,method='LSODA',t_eval=time_interval)

# The solution yH is y[0], so we assign the name yH to the solution y of the solve_ivp function which we named ''sol''
#The same with yH'[z] 
# which we call dotyH

yH=soly.y[0]
Y=soly.y[1]

#Now since the above are basically arrays, we switch the redshift to the initial definition by using the very own definition
# of the initial redshift-in our case recall we had to identify z=t in order for the solve_ivp function to work.

t=10-soly.t



#Now the definition of the Hubble rate of the present model defined as a function of the function yH and of the redshift.
#The outcome is an array, to be
#plotted as a function of the initially defined redshift which in terms of the programming dynamical variable t is t=10-soly.t
# look up above this comment.

def a(x):
    return 1/(11-x)

# Now we need to define some functions of the scale factor and of the functions yH and dotyH. These must be evaluated for the 
#initial time interval, not the redifed t, because the functions yH and dotyH were calculated in exactly this way. In the 
#plots we shall make use of the initial redshift values, but in the plots we basically connect values of arrays, 
#a list plot basically, so t=10-soly.t is just an array of values. However, the functions that will be built depending on yH and
#dotyH MUST be evaluated using the initial time interval, in our case the linspace time_interval

rs=(chi)*a(time_interval)**(-4)

H= np.sqrt(ms*(yH + a(time_interval)**(-3) + rs))

# The derivative of the Hubble rate
    
dHdz = (ms*(-3*(a(time_interval))**(-2) - 0.00124*a(time_interval)**(-3) + Y))/(2*np.sqrt(ms*(yH + a(time_interval)**(-3) + rs)))

Omega_Lambda = 0.6847

Omega_R = 8*10**(-5)

Omega_M = 0.3153 - Omega_R


#Dark energy density parameter


Omega_DE=yH/(yH+ a(time_interval)**(-3) + rs)


#Hubble Rate of LCDM

HLCDM=np.sqrt(H0p**2*(Omega_Lambda+ Omega_M* a(time_interval)**(-3) + Omega_R*a(time_interval)**(-4)))

#The derivative of the Hubble rate

dHLdz = -(H0p**2*(3*a(time_interval)**(-2)*Omega_M + 4*a(time_interval)**(-3)*Omega_R))/(2*np.sqrt(H0p**2*((a(time_interval))**(-3)*Omega_M+ (a(time_interval))**(-4)*Omega_R + Omega_Lambda)))


#Total Effective Equation of State parameter (Total EoS Parameter)

# For the $F(R)$ gravity model

weff=-1 - 2/3*a(time_interval)**(-1)*dHdz/H



#For the LCDM model

wefflcdm=-1 -2/3*a(time_interval)**(-1)*dHLdz/HLCDM;


#Dark Energy Equation of State Parameter:

wG=-1 -(1/3)*a(time_interval)**(-1)*Y/yH


#DECELERATION Parameter for the F(Rs) Gravity Model

q= -(a(time_interval)**(-1)/H)*dHdz - 1

#DECELERATION for the LCDM Model

qlcdm=-(a(time_interval)**(-1)/HLCDM)*dHLdz - 1
    
    
# Now the plots and the results of the numerical analysis
# The F(R)  model corresponds to blue color, and dashed-dot ''-.'' curve
#The LCDM model corresponds to red color and double dashed '--' curve.
    
#plt.style.use('Solarize_Light2')

plt.plot(time_interval,yH,marker='.',color='g',linestyle='-.',label='Programming Method')

plt.plot(t,yH,marker='.',color='b',linestyle='--',label='Correct Redshift Method')



plt.title('The statefinder $y_H$ vs Redshift' )


plt.xlabel('z')
plt.ylabel('$y_H$')

plt.tight_layout()

plt.grid(True)

plt.legend()

#plt.savefig('plot1.eps')

#plt.savefig('plot1.jpg')

plt.show()


plt.plot(time_interval,-Y,marker='.',color='g',linestyle='-.',label='Programming Method')

plt.plot(t,-Y,marker='.',color='b',linestyle='--',label='Correct Redshift Method')



plt.title('The statefinder derivative $y\'_H$ vs Redshift' )


plt.xlabel('z')
plt.ylabel('$y\'_H$')

plt.tight_layout()

plt.grid(True)

plt.legend()

#plt.savefig('dyHp.jpg')

plt.show()



plt.plot(t,H/H0p,marker='.',color='b',linestyle='-.',label='$f(R)$')
         
plt.plot(t,HLCDM/H0p,marker='.',color='r',linestyle='--',label='$\Lambda$CDM')


plt.title(' $H/H_0$ vs Redshift' )


plt.xlabel('z')
plt.ylabel('Hubble Rates')

plt.tight_layout()

plt.grid(True)

plt.legend()

#plt.savefig('Hubblep.jpg')

plt.show()


Ht=1.996*10**(-33)/H0p*np.ones(1000)

plt.plot(t,H/H0p,marker='.',color='b',linestyle='-.',label='$f(R)$')
         
plt.plot(t,HLCDM/H0p,marker='.',color='r',linestyle='--',label='$\Lambda$CDM')

plt.plot(t,Ht,marker='.',color='g',label='$H_t=1.996 10^{-33}eV$')


plt.title(' $H/H_0$ vs Redshift Close Up' )


plt.xlabel('z')
plt.ylabel('Hubble rates over $H_0$: H/H_0')

plt.tight_layout()

plt.grid(True)

plt.legend()

#plt.savefig('Hubblepc.jpg')


plt.ylim(1, 2)
plt.xlim(1.5,3)

plt.show()




    
plt.plot(t,Omega_DE,marker='.',color='b',linestyle='-.',label='$\Omega_{DE}(z)$-z for the f(R) Model')


plt.title('$\Omega_{DE}(z)$ vs Redshift' )


plt.xlabel('z')
plt.ylabel('$\Omega_{DE}(z)$')

plt.tight_layout()

plt.grid(True)

plt.legend()

#plt.savefig('Omdep.jpg')

plt.show()

         
plt.plot(t,weff,marker='.',color='b',linestyle='-.',label='$\omega_{eff}(z)$-z for the f(R) Model')
         
plt.plot(t,wefflcdm,marker='.',color='r',linestyle='--',label='$\omega_{eff}(z)$-z for the $\Lambda$CDM Model')



plt.title('Total EoS vs Redshift')


plt.xlabel('z')
plt.ylabel('EoS Parameters')

plt.tight_layout()

plt.grid(True)

plt.legend()

#plt.savefig('plot2.eps')

#plt.savefig('totaleosp.jpg')

plt.show()


plt.plot(t,wG,marker='.',color='b',linestyle='-.',label='$\omega_{DE}(z)$-z for the f(R) Model')



plt.title('Dark Energy EoS vs Redshift')


plt.xlabel('z')
plt.ylabel('EoS Parameters')

plt.tight_layout()

plt.grid(True)

plt.legend()

#plt.savefig('odep.jpg')

plt.show()

plt.plot(t,q,marker='.',color='b',linestyle='-.',label='$q$-z for the $f(R)$ Model')

plt.plot(t,qlcdm,marker='.',color='r',linestyle='-.',label='$q$-z for the $\Lambda$CDM Model')

plt.title('Deceleration Parameter vs Redshift')


plt.xlabel('z')
plt.ylabel('q')

plt.tight_layout()

plt.grid(True)

plt.legend()

#plt.savefig('plot3.eps')

#plt.savefig('decelerationp.jpg')

plt.show()




print("The dark energy density parameter $\Omega_DE(0)$ for the $f(R)$ Model at present time is:"+" "+str(Omega_DE[999]))
print() 
print() 
print("The dark energy EoS parameter for the $f(R)$ Model at present time is:"+" "+str(wG[999]))
print()  
print() 
print("The Hubble rate for the $f(R)$ Model at present time is:"+" "+str(H[999]))
print() 
print() 
print("The fraction of the Hubble rate for the $f(R)$ Model and of the $\Lmabda$CDM model at present time is:"+" "+str(H[999]/H0p))
print() 
print() 
print("The deceleration parameter for the $f(R)$ Model at present time is:"+" "+str(q[999]))
print()
```

<h2>Backwards Intergation Technique Using solve_ivp: Free fall and Another Illustrative Example</h2>

For our numerical analysis using the python 3 scipy module and the function solve_ivp of the scipy module, we made use of a backward integration technique, which we want to demonstrate how simple it is to work with, by using some simple examples. Also we shall demonstrate several tips for obtaining the correct results. Let us first consider a simple model of free fall motion in Earth's gravity field.

<h3>Free Fall</h3>

A ball of mass m=1kg is dropped from an altitude of x=10m above the earth to the ground. If the initial velocity of the ball is zero, find the position of the ball for all times, and the velocity before it touches the ground. $g=10 m/sec^{2}$.

This is a rather simple example to demonstrate how the solve_ivp module works. Obviously, from Newton's second law we have:

$$\vec{F}=m\frac{d^2\vec{x}}{dt^2}\, ,$$

with $F=mg$. The initial conditions are $x[0]=10m$, $x'[0]=v[0]=0$, and the analytic solution to the problem is: 

$$x[t]=-\frac{1}{2}gt^2+c_1t+c_2\, ,$$

where $c_1$ and $c_2$ are integration constants. By using the initial conditions, the solution is:

$$x[t]=10-\frac{1}{2}gt^2$$

so let us solve the differential equation numerically with Python by using solve_ivp:

First we need to transform it in the form $\frac{dy}{dt}=f(t,y)$, so we need to tranform the differential equation into an equivalent set of first order differential equations. This can be done easily:

set $v=\frac{dx}{dt}$ and the differential equation becomes,

$$v=\frac{d x}{d t}$$
$$\frac{d v}{d t}=-g$$,


or equivalently,
$$\frac{d x}{d t}=v$$
$$\frac{d v}{d t}=-g$$,

thus it is of the form $\frac{dy}{dt}=f(t,y)$, with $y=[x,v]$ and $f=[v,g]$. Now we can apply the solve_ivp method module of python. 

In the codes that will follow, we will solve the direct problem in an analytic and numerical way, and by using the results, we will solve the inverse problem.


First let us solve the direct problem of the free fall described above:

The ball is left from heigth $x_0$ at $t=0$ with initial velocity $v(0)=0$ and therefore the initial condition for $x$ is $x(0)=0$. We shall solve the equation for the first $t=\sqrt{2}$ seconds, since we know that at this point, the ball will have reached the ground. Then we shall solve the inverse problem, in which the initial conditions at $t=\sqrt{2}$ will be provided, and we will solve the problem numerically using a backwards integration technique, in order to overcome the onward integration procedure of the solve_ivp function.





```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

#Definition of the numerical solution for the problem

def f(t,y):
    g=10
    m=1
    v=y[0]
    x=y[1]
    dvdt=-g
    dxdt=v
    return np.array([dvdt,dxdt])

#final time instance
t_f=np.sqrt(2)

#duration of the motion, starting from zero and ending at t_f
t_span=np.array([0,t_f])
time_interval=np.linspace(t_span[0],t_span[1],10)

# The initial conditions, v, the velocity is initially zero at t=0, and x is equal to 10 at t=0.
y0=np.array([0,10])

#Now the solve_ivp function using LSODA
soly=solve_ivp(f,t_span,y0,method='LSODA',t_eval=time_interval)

v=soly.y[0]
x=soly.y[1]



t=soly.t

    
# The Analytic solution will also be presented for completeness and comparison with the numerical results.

def xp(t):
    g=10
    return -(1/2)*g*t**2+10

t_f=np.sqrt(2)

t1=np.linspace(0,t_f,10)

xs=xp(t1)

plt.plot(t1,xs,marker='d',color='b',linestyle='-',label='x-t Analytics')

plt.plot(t,x,marker='.',color='y',linestyle='-.',label='x-t Numerics')

plt.plot(t,v,marker='.',color='r',linestyle='-.',label='v-t Numerics')

plt.title('Position and Velocity (Red) vs Time Analytics (Green) and Numerical (Blue) Results')




plt.xlabel('t (sec)')
plt.ylabel('x,v')



plt.grid(True)

plt.legend()



plt.show()




```

Let us print the values of x and v at the end of the time interval, so at $t=\sqrt{2}$, which we shall use as feedback for the inverse problem we solve shortly:

```python
print('The value of x(t) at $t=\sqrt{2}$ is'+" "+"$x(\sqrt{2})$="+str(x[9]))
print('The value of v(t) at $t=\sqrt{2}$ is'+" "+"$v(\sqrt{2})$="+str(v[9]))
```

We shall use the above as initial conditions for the inverse problem. Now let us consider the backward integration technique.

<h3>Backward Integration Technique for the Free Fall</h3>


Now the inverse problem, suppose that we want to solve numerically the set of differential equations of the free fall, by assuming that the problem starts at $t=\sqrt{2}$sec and we use a counter clock and time decreases, so we want to find $x$ and $v$ at $t=0$, with the initial conditions at 
$t=\sqrt{2}$ being: 
$$x(\sqrt{2})=-9.99955673819386\times 10^{-11}$$
and
$$v(\sqrt{2})=-14.142135623730947$$
in appropriate physical units.

Hence, we want to solve the differential equations corresponding to the free fall but working backwards in time from $t=\sqrt{2}$ to $t=0$. However the solve_ivp function only works forward in time, thus we have to use a trick.

We choose the variable $t'=\sqrt{2}-t$, hence the original differential equation becomes,
$$\frac{d^2x}{dt'^2}=-g\, ,$$

and the system of ODEs becomes,

$$\frac{d x}{d t'}=v$$
$$\frac{d v}{d t'}=-g$$


and also the initial conditions in the new $t'$ system are: 


$t'=0$ ($t=\sqrt{2}$) $\,\,\,\,\,x(t'=0)=-9.99955673819386\times 10^{-11}$


$t'=0$ ($t=\sqrt{2}$) $\,\,\,\,\,v(t'=0)=14.142135623730947$

Note here the change in sign for the velocity, this is due to the fact that $\frac{d v}{d t'}=-\frac{d v}{d t}$. We have to take this change into account later on when we will want to plot the velocity as a function of the time, and of course this change of sign must be used in any velocity dependent quantity.


The new problem is easy to solve numerically though, it is basically the same problem, so the results are below and the numerical analysis, the plots are $x-t$ in the $t$ system.  To better explain the switch to the original time, the solve_ivp will return x and v as arrays corresponding to the following $t'$ array:
$$t'=[t'_{0},t'_{1},....,t'_{9}],\,\,\,with\,\,\,t'_0=0,\,\,\,t'_{9}=\sqrt{2}$$ 
which array corresponds to the following $t$ array,
$$t=[t_{9},t_{8},....,t_{1},t_0],\,\,\,with\,\,\,t_0=10,\,\,\,t_{9}=0$$. 

Now the outcome of the solve_ivp $t'$ integration is basically $x$ and $v$, which in the $t$ coordinate are $x$ and $-v$ (recall $\frac{d v}{d t'}=-\frac{d v}{d t}$).

What we expect to find by this inverse integration is $x(t'=\sqrt{2})=x(t=0)\simeq 10$ and $-v(t'=\sqrt{2})=v(t=0)\simeq 0$.

Let us run the code and we plot the results and return the values of $x$ and $v$ at $t'=\sqrt{2}$. Note that the plots are $x-t$ and $-v-t$ and $t$ is an array $t=\sqrt{2}-t'$.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def f(t,y):
    g=10
    m=1
    v=y[0]
    x=y[1]
    dvdt=-g
    dxdt=v
    return np.array([dvdt,dxdt])

t_span=np.array([0,np.sqrt(2)])
time_interval=np.linspace(t_span[0],t_span[1],10)

y0=np.array([14.142135623730947,-9.99955673819386e-11])



soly=solve_ivp(f,t_span,y0,method='LSODA',t_eval=time_interval)

v=soly.y[0]
x=soly.y[1]


t=np.sqrt(2)-soly.t

    

plt.plot(t,x,marker='d',color='#000000',linestyle='--',label='x-t Numerics')


plt.plot(t,-v,marker='d',color='r',linestyle='--',label='v-t Numerics')

plt.title('Position and Velocity Analytics and Numerical Results')




plt.xlabel('t (sec)')
plt.ylabel('x')

plt.tight_layout()

plt.grid(True)

plt.legend()

print("The value of x at t'=\sqrt{2} or t=0 is: x= "+str(x[9]))
print("The value of v at t'=\sqrt{2} or t=0 is: v= "+str(-v[9]))
```

So the numerical results are very close to the values we expected.

<h2>Another Illustrative Example</h2>

The previous problem was kind of easy because the differential equation was not changed under the $t$ transformation. Now we shall present a somewhat more interesting problem.


Consider the following differential equation,
$$\frac{d^2x}{dt^2}-\frac{1}{t}\frac{dx}{dt}=0$$
with the corresponding initial conditions being,
$$x(0)=10.5,\,\,\,v(0)=1,$$
where $v(t)=x'(t)$. This differential equation can be solved numerically and analytically so let us show the analytic solution first, and then we proceed with the numerical results. Suppose we aim to find the values of $x(t)$ and $v(t)$ for the time interval $t=[1,5]$.  

The analytic solution is:
$$x(t)=10+0.5t^2\, ,$$
$$v(t)=t\, ,$$
so the values of $x$ and $v$ at $t=5$sec are $x(5)=22.5$ and $v(5)=5$ in appropriate units. Now we aim to find the behavior of $x$ and $v$ numerically and in addition to find the values of $x$ and $v$ at $t=5$sec, to see whether the numerical and analytical solutions coincide. The inverse problem will be based on the values of $x$ and $v$ at $t=5$sec. Below we present the plots of the numerical and analytic solutions, and also the numerical values of $x$ and $v$ at $t=5$sec, using the solve_ivp function of the SCIPY module, and the method used is LSODA. Of course, the differential equation is transformed to the following set of ODEs:
$$\frac{dx}{dt}=v$$
$$\frac{dv}{dt}=\frac{1}{t}v$$
and this is vital to transform the original differential equation to the given form of two ODEs in order for the solve_ivp function to be able to perform the numerical analysis.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def f(t,y):
    v=y[0]
    x=y[1]
    dvdt=(1/t)*v
    dxdt=v
    return np.array([dvdt,dxdt])

t_span=np.array([1,5])
time_interval=np.linspace(t_span[0],t_span[1],10)

y0=np.array([1,10.5])




soly=solve_ivp(f,t_span,y0,method='LSODA',t_eval=time_interval)

v=soly.y[0]
x=soly.y[1]


t=soly.t


def x1(t2):
    return 0.5*t2**2+10

def v1(t2):
    return t2

t1=np.linspace(1,5)

plt.plot(t,x,marker='H',color='r',linestyle='--',label='x-t Numerics')

plt.plot(t,v,marker='H',color='b',linestyle='-.',label='v-t Numerics')

plt.plot(t1,x1(t1),marker='.',color='y',linestyle='--',label='x-t Analytic')

plt.plot(t1,v1(t1),marker='.',color='y',linestyle='-.',label='v-t Analytic')



plt.title('Position and Velocity Analytics and Numerical Results')



plt.xlabel('t (sec)')
plt.ylabel('x,v')

plt.tight_layout()

plt.grid(True)

plt.legend()

print("The position at t=5sec is x(5)="+str(x[9]))
print("The velocity at t=5sec is v(5)="+str(v[9]))
```

As it can be seen from the results, the the analytic and numerical solutions overlap, and the analytic solution is excellently approached by the numerical solution. Now let us proceed with the inverse problem which we now describe.

<h3>Backward Integration Problem</h3>

Now we wish to solve the same differential equation,
$$\frac{d^2x}{dt^2}-\frac{1}{t}\frac{dx}{dt}=0$$
but this time we have the following initial conditions,
$$x(5)=22.5,\,\,\,v(5)=5,$$
and we aim to integrate backwards the set of ODE's and find the values of $x$ and $v$ at $t=1$sec, so backward in time. According to the previous problem, and of course by taking into account the analytic solutions, these are:
$$x(1)=10.5,\,\,\,v(1)=1$$
in appropriate units.

Thus we need to find a technique to trick the solve_ivp function and integrate it onward for some appropriate variable. We use the transformation:
$$t'=6-t\, ,$$
thus the derivatives are transformed as,
$$\frac{d}{dt'}=-\frac{d}{dt}$$
hence the initial differential equation becomes,
$$\frac{d^2x}{dt'^2}+\frac{1}{6-t'}\frac{dx}{dt'}=0\, ,$$
hence by setting,
$$\frac{dx}{dt'}=v$$
we transform the differential equation to a set of ODE's,
$$\frac{dx}{dt'}=v$$
$$\frac{dv}{dt'}=-\frac{1}{6-t'}v$$
while the initial conditions are,
$$x(5)=22.5,\,\,\,v(5)=5,$$ 
which in the new coordinate $t'$ are transformed as,
$$x(1)=22.5,\,\,\,v(1)=-5,$$
since $v=\frac{dx}{dt'}=-\frac{dx}{dt}$.
Thus we converted the problem of backward integration to a forward integration problem, and now we shall solve it numerically using the solve_ivp function of the SCIPY module.

We aim to find the same curves as before, and also the values of $x$ and $v$ at the original time $t=1$ to be,
$$x(1)=10.5,\,\,\,v(1)=1$$
in appropriate units.
Note that, in order to make contact with the origina time coordinate $t$, after solving numerically the system of ODE's for the $t'$ coordinate, we must remember that $\frac{dx}{dt}=-v$, where $v$ is the result of the numerical integration in the $t'$ coordinate, so it is an array.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def f(t,y):
    g=10
    m=c=1
    v=y[0]
    x=y[1]
    dvdt=-(1/(6-t))*v
    dxdt=v
    return np.array([dvdt,dxdt])

t_span=np.array([1,5])
time_interval=np.linspace(t_span[0],t_span[1],10)

y0=np.array([-5,22.5])



soly=solve_ivp(f,t_span,y0,method='LSODA',t_eval=time_interval,dense_output=True)

v=soly.y[0]
x=soly.y[1]

t=6-soly.t
    

plt.plot(t,x,marker='d',color='#000000',linestyle='--',label='x-t Numerics')


plt.plot(t,-v,marker='d',color='r',linestyle='--',label='v-t Numerics')

plt.title('Position and Velocity Numerical Results, Backward Integration')




plt.xlabel('t (sec)')
plt.ylabel('x,v')

plt.tight_layout()

plt.grid(True)

plt.legend()

print("The position at t=5sec is x(5)="+str(x[9]))
print("The velocity at t=5sec is v(5)="+str(-v[9]))
```

As it can be seen, the outcome of the integration is exactly the one we expected, thus this demonstrates the accuracy of the backwards integration technique we used. But let us also compare the plots of the analytic solution with the numerical which we just obtained. We expect exact coincidence. The results are quoted below.
